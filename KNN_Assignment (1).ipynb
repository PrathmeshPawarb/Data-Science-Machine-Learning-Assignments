```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
zoo = pd.read_csv('/content/Zoo.csv')
zoo.head()
```

```python
df = zoo.iloc[:,1:]
df.head()
```

```python
df.isna().sum()
```

```python
df.describe()
```

As we see legs column not contain binary data


```python
df['has_legs'] = np.where(df['legs']>0,1,0)
df = df[['hair','feathers','eggs','milk', 'airborne', 'aquatic', 'predator', 'toothed', 'backbone', 'breathes','venomous','fins','legs','has_legs','tail','domestic','catsize','type']]
df.head()
```

```python
df.drop('legs', axis=1, inplace=True)
df.head()
```

```python
sns.countplot(x='type', data=df)
```

```python
df['type'].unique()
```

```python
#spliting data

x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

xtest,xtrain,ytest,ytrain = train_test_split(x, y, test_size=0.30, random_state=5)
```

#visualization

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_values = [i for i in range (1,15)]
scores = []



for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, xtrain, ytrain, cv=5)
    scores.append(np.mean(score))

scores
```

```python
plt.plot(scores)
```

There is significant change in accuracy after tranforming legs column into binary.

```python
# KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain,ytrain)

knn.score(xtest,ytest)
```

```python
# Prediction

ypred = knn.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred,ytest)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(ytest,ypred)
cm
```

```python
sns.heatmap(cm, annot= True)
```

```python
plt.figure(figsize=(10,10))
plt.hist(ytest,label = 'actual',color = 'm')
plt.hist(ypred,label = 'prediction',color = 'c')
'''ax.hist(ytest, color = 'm', alpha = 0.5, label = 'actual', bins=7)
ax.hist(ypred, color = 'c', alpha = 0.5, label = 'prediction', bins=7)
ax.yaxis.set_ticks(np.arange(0,11))
ax.legend(loc = 'best')
plt.show()
'''
```

#Glass

```python
glass = pd.read_csv('/content/glass.csv')
glass.head()
```

```python
glass.info()
```

```python
glass['Type'].unique()
```

```python
glass.describe()
```

```python
# splitting into dependent and independent variable

x1 = glass.iloc[:,:-1]
y1 = glass.iloc[:,-1]
```

```python
def norm_func(i):
    x = (i-i.min())	/ (i.max()-i.min())
    return (x)

x1 = norm_func(x1)
```

```python
x1
```

```python
#splitting Test train data

from sklearn.model_selection import train_test_split

x1test,x1train,y1test,y1train = train_test_split(x1, y1, test_size=0.20, random_state=5)
```

```python
# Grid Search for Algorithm Tuning
from sklearn.model_selection import GridSearchCV

n = np.array(range(1,31))                               #Creating array for k value from 1 to 51.
parameter = dict(n_neighbors=n)                         #Creating Dictonary containing these k values.


model = KNeighborsClassifier()                           # Using GridSearchCV with model=(knn)and parameter
grid = GridSearchCV(estimator=model, param_grid=parameter)
grid.fit(x1train, y1train)
```

```python
print(grid.best_score_)             #Score
print(grid.best_params_)            #for cluster nos.
```

```python
m = KNeighborsClassifier(n_neighbors = 8)
m.fit(x1train, y1train)



from sklearn.metrics import accuracy_score
ypred1 = m.predict(x1test)
print('Test_accuracy=', accuracy_score(y1test, ypred1))

from sklearn.metrics import accuracy_score
ypred2 = m.predict(x1train)
print('Train_accuracy=', accuracy_score(y1train, ypred2))
```
