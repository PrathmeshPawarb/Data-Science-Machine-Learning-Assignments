```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv('/content/book.csv', encoding='latin-1')
book
```

```python
book.columns
```

```python
df= book.iloc[:,1:]
df.rename(columns={'User.ID':'ID','Book.Title':'Book','Book.Rating':'Rating'}, inplace=True)
df
```

```python
df['ID'].nunique()
```

```python
df['Book'].nunique()
```

```python
df['Rating'].nunique()
```

```python
df.sort_values('ID')
```

```python
newtable = df.pivot_table(index="ID", columns="Book",values="Rating",aggfunc='sum')
newtable
```

```python
newtable.fillna(0, inplace=True)
newtable
```

```python
from sklearn.metrics import pairwise_distances
```

```python
sim_user = 1 - pairwise_distances(newtable.values, metric="cosine")
```

```python
sim_user
```

```python
sim_user_df = pd.DataFrame(sim_user)
sim_user_df
```

```python
# chenging index of matris by user id

sim_user_df.columns = book['User.ID'].unique()
sim_user_df.index = book['User.ID'].unique()

```

```python
sim_user_df.iloc[0:5, 0:5]
```

```python
#filling digonal values to zero
np.fill_diagonal(sim_user, 0)
sim_user_df.iloc[0:5, 0:5]
```
